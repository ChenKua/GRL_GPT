```python
import pandas as pd
```

```python
xls = pd.ExcelFile('result.xlsx') # read excel
R1 = pd.read_excel(xls, 'R1')   # read specific sheet
R2 = pd.read_excel(xls, 'R2')
```

# Kendell

```python
# https://www.geeksforgeeks.org/python-kendall-rank-correlation-coefficient/
# Import required libraries
from scipy.stats import kendalltau

# Round 1: E1-1

for i in range(4):
  index1 = i * 2
  index2 = index1 + 1
# Taking values from the above example in Lists
  X = R1.iloc[:,index1].tolist() 
  Y = R1.iloc[:,index2].tolist()
  
  # Calculating Kendall Rank correlation
  corr, _ = kendalltau(X, Y)
  print('Kendall Rank correlation: %.5f' % corr)
    
  # This code is contributed by Amiya Rout
```

```python
from statsmodels.stats.inter_rater import aggregate_raters as aggregate_raters
from statsmodels.stats.inter_rater import fleiss_kappa
import numpy as np
```

```python
# https://www.geeksforgeeks.org/python-kendall-rank-correlation-coefficient/
# Import required libraries
from scipy.stats import kendalltau

# Round 1: E1-1

for i in range(4):
  index1 = i * 2
  index2 = index1 + 1
# Taking values from the above example in Lists
  X = R1.iloc[:,index1].tolist() 
  Y = R1.iloc[:,index2].tolist()
  
  data = np.column_stack((X, Y))
  # print(data.shape)
  table = aggregate_raters(data)
  kappa = fleiss_kappa(table[0], method='fleiss')
  print(kappa)
```
