```python
# from langchain.llms import OpenAI
# from langchain.chat_models import ChatOpenAI
import pandas as pd
import openai
```

```python
# https://platform.openai.com/docs/models
# model_name = gpt-4, gpt-3.5-turbo
# llm = ChatOpenAI(temperature=1, model_name="gpt-4")
# https://python.langchain.com/en/latest/modules/models/chat/integrations/openai.html
```

```python
# llm("hello")
```

```python
openai.api_key = openai_key

def run_llm(prompt):  
  completion = openai.ChatCompletion.create(
      temperature = 1,
      model="gpt-3.5-turbo", # gpt-3.5-turbo, gpt-4
      messages=[
              {"role": "user", "content": prompt},                
          ]
      )
  return(completion['choices'][0]['message']['content'])
```

```python
size = experiment_1.shape[0]

result = ["placeholder" for i in range(size)]
```

```python
experiment_1_result.iloc[0,4] = "sdss"
```

```python
context = """You are a software engineering student on an exam for goal-oriented requirement engineering. 
Follow the instructions in the question and answer the following question concisely.


"""

for index, prompt in enumerate(experiment_1['content']):
  prompt = context + prompt
  result = run_llm(prompt)
  print(index)
  print(result)
  experiment_1_result.iloc[index,4] = result

  # save in every run
  experiment_1_result.to_excel("3K-3.xlsx")
```

```python
for index, prompt in enumerate(experiment_1['content']):
  if index >= 15:
    result = run_llm(prompt)
    print(index)
    print(result)
    experiment_1_result.iloc[index,4] = result

    # save in every run
    experiment_1_result.to_excel("3K-3.xlsx")
```
